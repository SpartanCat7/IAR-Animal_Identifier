```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm

import tensorflow as tf
from tensorflow import keras
```

```python
DATADIR = "PhotosToProcess"

CATEGORIES = ["CAT", "CHICKEN", "DOG", "HORSE"]
IMG_SIZE = 100
```

```python
original_photos = []

def LoadData():

    for img in tqdm(os.listdir(DATADIR)):
        try:
            img_array = cv2.imread(os.path.join(DATADIR,img) ,cv2.IMREAD_GRAYSCALE)
            new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
            original_photos.append(new_array)
        except Exception as e: 
            pass
```

```python
LoadData()

data_to_process = np.array(original_photos).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
data_to_process = data_to_process/255
```

```python
MODELS_DIR = 'SavedModels/'

loaded_models = []

for category in CATEGORIES:
    model = keras.models.load_model(MODELS_DIR + category)
    loaded_models.append([model, category, 0]) # 0 will next be replaced with the predictions of the model
```

```python
for model, category, predictions in loaded_models:
    index = loaded_models.index([model, category, predictions])
    prediction = model.predict(data_to_process)
    loaded_models[index][2] = prediction
    print(category)
    print(prediction)
```

```python
index = 0
for item in original_photos:
    plt.imshow(item, cmap='gray')
    plt.show()

    item_predictions = []

    for model, category, predictions in loaded_models:
        item_predictions.append([category, predictions[index]])

    for category, prediction in item_predictions:
        print(category + ': ' + str(prediction))
    index += 1
```
